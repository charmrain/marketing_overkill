```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    name_pattern = re.compile(r'<title>(.*?)</title>')

    # Find all email addresses and names in the HTML
    emails = re.findall(email_pattern, html_content)
    names = re.findall(name_pattern, str(soup.title))

    return emails, names

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, names = extract_email_and_name(url)

print("Emails:", emails)
print("Names:", names)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    name_pattern = re.compile(r'Contact: (.+?)\n')

    # Find all email addresses using the regular expression
    emails = re.findall(email_pattern, html_content)

    # Find the name using BeautifulSoup and the name_pattern
    name_match = re.search(name_pattern, html_content)
    name = name_match.group(1) if name_match else None

    return emails, name

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, name = extract_email_and_name(url)

print("Name:", name)
print("Emails:", emails)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    contact_pattern = re.compile(r'Contact: (.+?)$')

    # Find all email addresses and contact names in the HTML
    emails = re.findall(email_pattern, html_content)
    contact_names = [match.group(1) for match in re.finditer(contact_pattern, html_content)]

    return emails, contact_names

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, contact_names = extract_email_and_name(url)

print("Emails:", emails)
print("Contact Names:", contact_names)
```

```python
from bs4 import BeautifulSoup

html_content = """
<div class="company-information">
    <p><br/></p><span style="font-weight: bold; color: red; font-size: 1.8em;">180 Degrees Inc</span><br/>236 Clifton Ave<br/>Minneapolis, MN 55403-3466<br/><span style="font-weight: bold;">Phone Number</span>: 612-813-5010<br/><br/><p><br/></p><span style="font-weight: bold;">Number of Employed</span>: 100<br/><span style="font-weight: bold;">Annual Revenues</span>: $5<br/><span style="font-weight: bold;">Corporate Email:</span>: <br/><span style="font-weight: bold;">Business Start Date</span>: 1971<br/><p><br/></p><span style="font-weight: bold;">Primary Contact </span>: Richard Gardell<br/><span style="font-weight: bold;">Primary Email</span>: Richard@180degrees.org<br/><p><br/></p><span style="font-weight: bold;">Secondary Contact</span>: Richard Gardell<br/><span style="font-weight: bold;">Secondary Email</span>: Richard@180degrees.org<br/><p><br/></p><span style="font-weight: bold;">Tertiary Contact</span>: Richard Gardell<br/><span style="font-weight: bold;">Tertiary Email</span>: Richard@180degrees.org<br/><span style="font-weight: bold;">Tertiary Phone</span>: 612-813-5010 <br/><p><br/></p><span style="font-weight: bold;">Quatemary Contact</span>: Richard Gardell<br/><span style="font-weight: bold;">Quatemary Email</span>: Richard@180degrees.org<br/><span style="font-weight: bold;">Quatemary Phone</span>: 612-813-5010 <br/><p><br/></p><span style="font-weight: bold;">Quinary Contact Person</span>: Sarah Walker<br/><span style="font-weight: bold;">Quinary Email</span>: Sarah@180degrees.org<br/><span style="font-weight: bold;">Quinary Phone</span>: 612-813-5017 <br/><p><br/></p><p><br/></p><span style="font-weight: bold;">Categories of Service</span><br/><p>Other Individual and Family Services </p>
    <p><br/></p>
</div>
"""

soup = BeautifulSoup(html_content, 'html.parser')

# Extract company name, address, and phone number
company_name = soup.find('span', style="font-weight: bold; color: red; font-size: 1.8em;").text.strip()
address = soup.find_all('br')[2].next_sibling.strip()
phone_number = soup.find('span', style="font-weight: bold;").next_sibling.strip()

# Extract number of employed, annual revenues, corporate email, and business start date
employed = soup.find('span', style="font-weight: bold;").find_next('span').text.strip()
annual_revenues = soup.find('span', style="font-weight: bold;").find_next('span').find_next('span').text.strip()
corporate_email = soup.find('span', style="font-weight: bold;").find_next('span').find_next('span').find_next('span').next_sibling.strip()
business_start_date = soup.find('span', style="font-weight: bold;").find_next('span').find_next('span').find_next('span').find_next('span').next_sibling.strip()

# Extract primary, secondary, tertiary, quatemary, and quinary contacts
contacts = soup.find_all('span', style="font-weight: bold;")
contact_info = {}
for contact in contacts:
    contact_type = contact.text.strip().replace(' ', '_').lower()
    contact_name = contact.find_next('span').text.strip()
    contact_email = contact.find_next('span').find_next('span').text.strip()
    contact_info[contact_type + '_name'] = contact_name
    contact_info[contact_type + '_email'] = contact_email

# Extract categories of service
categories_of_service = soup.find('span', style="font-weight: bold;").find_next('p').text.strip()

# Print the extracted information
print("Company Name:", company_name)
print("Address:", address)
print("Phone Number:", phone_number)
print("Number of Employed:", employed)
print("Annual Revenues:", annual_revenues)
print("Corporate Email:", corporate_email)
print("Business Start Date:", business_start_date)

for contact_type in ['primary', 'secondary', 'tertiary', 'quatemary', 'quinary']:
    print(f"{contact_type.capitalize()} Contact Name:", contact_info.get(f"{contact_type}_name", ""))
    print(f"{contact_type.capitalize()} Contact Email:", contact_info.get(f"{contact_type}_email", ""))

print("Categories of Service:", categories_of_service)
```
