```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    name_pattern = re.compile(r'<title>(.*?)</title>')

    # Find all email addresses and names in the HTML
    emails = re.findall(email_pattern, html_content)
    names = re.findall(name_pattern, str(soup.title))

    return emails, names

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, names = extract_email_and_name(url)

print("Emails:", emails)
print("Names:", names)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    name_pattern = re.compile(r'Contact: (.+?)\n')

    # Find all email addresses using the regular expression
    emails = re.findall(email_pattern, html_content)

    # Find the name using BeautifulSoup and the name_pattern
    name_match = re.search(name_pattern, html_content)
    name = name_match.group(1) if name_match else None

    return emails, name

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, name = extract_email_and_name(url)

print("Name:", name)
print("Emails:", emails)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def extract_email_and_name(url):
    # Send an HTTP request to the URL and get the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Use regular expressions to find email addresses and names
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    contact_pattern = re.compile(r'Contact: (.+?)$')

    # Find all email addresses and contact names in the HTML
    emails = re.findall(email_pattern, html_content)
    contact_names = [match.group(1) for match in re.finditer(contact_pattern, html_content)]

    return emails, contact_names

# Example usage
url = "https://fastcashconsulting.com/company/180-degrees-inc-6128135010"
emails, contact_names = extract_email_and_name(url)

print("Emails:", emails)
print("Contact Names:", contact_names)
```
